```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/test_energy_data (1).csv")
```

```python
data.isnull().sum()
```

```python
mean_val =  data['Energy Consumption'].mean()
median_val = data['Energy Consumption'].median()
mode_val = data['Energy Consumption'].mode()[0]

print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
```

```python
print(data['Building Type'].value_counts())
print('=========================================')
print(data['Day of Week'].value_counts())
```

```python
data = pd.get_dummies(data, columns=['Building Type', 'Day of Week'], drop_first=True, dtype=int)
data.head()
```

```python
x = data.drop('Energy Consumption', axis=1)
y = data['Energy Consumption']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
r2_score = model.score(X_test, y_test)
mse = metrics.root_mean_squared_error(y_test, y_pred)

print(f'R-squared: {r2_score}')
print(f'Root Mean Squared Error: {mse}')
```

```python
import pickle

# Save the preprocessed data (including encoded 'Building Type' and 'Day of Week')
with open('preprocessed_energy_data.pkl', 'wb') as f:
    pickle.dump(data, f)
print("Preprocessed data saved to 'preprocessed_energy_data.pkl'")


# Save the trained model
with open('linear_regression_model.pkl', 'wb') as f:
    pickle.dump(model, f)
print("Trained Linear Regression model saved to 'linear_regression_model.pkl'")
```
